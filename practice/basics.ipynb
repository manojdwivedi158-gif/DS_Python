```python
# leap year program

year=int(input("enter the year:"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year," this is leap year")
        else:
            print("not a leap year")
    else:
        print(year,"leap year")
    
else:

  print("not leap year")    
```

```python
# while loop

# while loop continues the execution until the condition fails   or  continues the execution until condition fails


num=int(input("enter your number"))
while num<12:
    print(num)
    num=num+1
```

```python
num1=int(input("enter your number:"))
num2=int(input("enter your number"))
operation=input("enter the operation")
if operation=="+":
    print(num1+num2)
elif operation=="-":
    print(num1-num2)
elif operation=="*":
    print(num1*num2)
elif operation=="/":
    print(num1/num2)
elif operation=="//":
    print("floor division ",num1//num2)
else:
    print("operation is unidentified")
```
