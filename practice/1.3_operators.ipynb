```python
# not equal to 
# greater then
#less then

val1=5
val2=10
print(val1 != val2)
print(val1 > val2)  # 5 is not greater then 10

print(val1 < val2) # 10 is greater then 5

print(val1 >= val2)  #5 is not greater then or equal to 10
print(val1 <= val2)  #5 is less then 10

print(val1 // val2) # 5 divided by 10
print(val2 // val1) # 10 divided by 5
print(val1 ** val2)  # 5 ki power 10
```
