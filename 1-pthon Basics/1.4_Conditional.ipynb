```python
# Write a program to determine leep year

year=int(input("enter the year:"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year," is a leap year")
        else:
            print(year," is not a leap year")
        

    else:
        print(year," is a leap year")
else:
    print(year," is not a leap year")
```

```python
# simple calculator program:

num1=int(input("enter your number:"))
num2=int(input("enter your second number"))
operation=input("enter the operation +,-,*,/,%")
if operation == "+":
    print(num1+num2)

elif operation == "-":
    print(num1-num2)
elif operation == "*":
    print(num1*num2)
elif operation == "/":
    print(num1/num2)
elif operation == "%":
    print(num1%num2)

else:
    print("invalid operation")
```
